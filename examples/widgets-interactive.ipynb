# 交互式 widget

基于 widget interactive 特性，我们可以通过 slider 等 widget 控件来刷新 matplotlib 输出的图表

```python
import ipywidgets as widgets
import matplotlib.pyplot as plt 
import numpy as np

def plot_func(freq):
    x = np.linspace(0, 2 * np.pi)
    y = np.sin(freq * x)
    plt.plot(x,y)
    plt.ylim(-1,1)
    plt.show()

widgets.interactive(plot_func, freq=widgets.IntSlider(value=2, min=1, max=10))
```
